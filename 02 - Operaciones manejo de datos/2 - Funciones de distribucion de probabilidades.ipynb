# Funciones de distribución de probabilidades

## distribución Uniforme

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
a = 1
b = 100
n = 1000000
data = np.random.uniform(a, b, n)
```

```python
%matplotlib inline
plt.hist(data)
```

## Distribución Normal

```python
data = np.random.randn(1000000)
```

```python
#data
```

```python
x = range(1, 1000001)
plt.plot(x, data)
```

```python
%matplotlib inline
plt.hist(data)
```

```python
plt.plot(x, sorted(data))
```

```python
mu = 5.5
sd = 2.5
z_10000 = np.random.randn(10000)
data = mu + sd * z_10000  # z (distribucion normal) = (X - mu) / sd  -> N(0, 1), X = mu + sd * Z
plt.hist(data)
```
