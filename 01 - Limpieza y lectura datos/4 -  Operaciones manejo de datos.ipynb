# Operaciones manejo de datos
## Data Wrangling - La cirugía de los datos

El **data wrangling**, aveces denominada data munging, es el proceso de transformar y mapear datos de un dataset raw (en bruto) en otro formato con la intencion de hacerlo mas apropiado y valioso para una variedad de propositos posteriores, como el analisis. Un **data wrangler** es una persona que realiza estas operaciones de transformacion.

Esto puedo incluir **munging**, visualizacion de datos, agregacion de datos, entrenamiento de un modelo estadistico, asi como muchos otros usos potenciales. La oscilacion de datos como proceso generalmente sigue un conjunto de pasos generales que comienzan extrayendo los datos en forma cruda del origien de datos, dividiendo los datos en bruto usando algoritmos (por ejemplo clasificacion) o analizando los datos en estructuras de datos predefinidos, y finalmente depositando el contenido en un sistema de almacenamiento para su futuro uso.

**Cuando se selecciona una sola columna se convierte en Series y multiples columnas DataFrame**

```python
desired_columns = ['Account Length', 'Phone', 'Eve Charge', 'Day Calls']
subset = data[desired_columns]

subset.head()
```

```python
desired_columns = ['Account Length', 'Email Message', 'Day Calls']
desired_columns
```

```python
all_columns_list = data.columns.values.tolist()
all_columns_list
```

**Usando sublist**

```python
sublist = [x for x in all_columns_list if x not in desired_columns]
```

```python
subset = data[sublist]
subset
```

### Varias condiciones
#### AND : &

```python
data3 = data[(data['Day Mins'] > 300) & (data['State'] == 'NY')]
data3
```

#### OR : |

```python
data4 = data[(data['Day Mins'] > 300) | (data['State'] == 'NY')]
data4.shape
```
