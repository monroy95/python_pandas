## Parametros read_csv

**Ejemplo** 
```python
read.csv(filepath='../datasets/titanic/titanic3.csv', sep=',', dtype={'Ingresos':np.float64, 'edad':np.int32},
         header=0, names={'Ingresos', 'edad'}, skiprows=12, index_col=None, skip_blank_lines=False, na_filter=False)
```
* **sep**: podemos usar expresiones regulares, por default casisiempre es , comma
* **dtype**: podemos formatear una columna a un tipo de dato, (fechas, numeros, float... etc) np = Numpy
* **header**: Por default esta en none = la primera fila es cabecera, puede ser un numero entero o una lista, especifica que fila se usaran de cabeceras
* **names**: sirve para indicar los nombres de las columnas, creamos una lista con los valores
* **skiprows**: puede ser un entero o una lista, sirve para saltarse determinado numero de filas
* **index_col**: por default es none, podemos indicar que columna puede ser la etiqueta de cada fila = index
* **skip_blank_lines**: podemos saltar las lineas en blanco, True/False
* **na_filter**: Detecta los marcadores de valores que falten dentro del dataset, na, string vacios, elimina aquellas filas que contengan valores desconocidos

## Carga de datos a traves de la funcion open

````python
data3 = open(mainpath + 'customer-churn-model/Customer Churn Model.txt', 'r')
````

````python
# strip: elimina especio al inicio y final
# split usada para dividir
cols = data3.readline().strip().split(',')
n_cols = len(cols)
````

````python
counter = 0
main_dict = {}
for col in cols:
    main_dict[col] = []
````

````python
for line in data3:
    values = line.strip().split(',')
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    counter += 1
    
print('El dataset tiene %d filas y %d columnas'%(counter, n_cols))
````

````python
## Cargamso la data del archivo a un dataframe
df3 = pd.DataFrame(main_dict)
df3.head()
````
